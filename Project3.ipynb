# Topic 2) Crimes Across the Years:

    Is the total number of crimes increasing or decreasing across the years?
    Are there any individual crimes that are doing the opposite (e.g., decreasing when overall crime is increasing or vice-versa)?

```python
crimes_per_year = df.resample('Y').size().reset_index(name='Total Crimes')
crimes_per_year['Difference'] = crimes_per_year['Total Crimes'].diff()
# Percentage of increase/decrease from previous year
crimes_per_year['Percentage Change'] = (crimes_per_year['Difference'] / crimes_per_year['Total Crimes']) * 100
total_crimes_2001 = crimes_per_year.loc[crimes_per_year['Date'].dt.year == 2001, 'Total Crimes'].values[0]
crimes_per_year['Change since 2021'] = ((crimes_per_year['Total Crimes'] - total_crimes_2001) / total_crimes_2001) * 100

crimes_per_year
```

```python
crimes_by_type_per_year = df.groupby([df.index.year, 'Primary Type']).size().reset_index(name='Total Crimes')
crimes_by_type_per_year.columns = ['Year', 'Primary Type', 'Total Crimes']

crimes_by_type_per_year
```

```python
crimes_2001 = crimes_by_type_per_year[crimes_by_type_per_year['Year'] == 2001]
crimes_2022 = crimes_by_type_per_year[crimes_by_type_per_year['Year'] == 2022]
crimes_comparison = pd.merge(crimes_2001, crimes_2022, on='Primary Type', suffixes=('_2001', '_2022'))
crimes_comparison.set_index('Primary Type', inplace=True)
crimes_comparison['Percentage Change'] = ((crimes_comparison['Total Crimes_2022'] - crimes_comparison['Total Crimes_2001']) / crimes_comparison['Total Crimes_2001']) * 100
crimes_comparison.sort_values('Percentage Change', ascending=False)
```

### While gambling and prostitution are almost completely gone, concealed carry license violations have risen by an incredible amount of 17,400%, and criminal sexual assault by 5268%, aswell as some other crimes with a smaller percentage such as Obscenity, stalking and weapons violations doubling from 2001 to 2022

# Topic 4) Comparing Months:

    ​What months have the most crime? What months have the least?
    Are there any individual crimes that do not follow this pattern? If so, which crimes?

```python
total_crimes_per_month = (df.resample('M').size()
    .reset_index(name='Total Crimes')
    .assign(Month=lambda x: x['Date'].dt.month)
    .drop(columns='Date')
    .groupby('Month')
    .sum()
    .sort_values(by='Total Crimes', ascending=False)
)
total_crimes_per_month
```

### Most crimes were committed on July and August, while the lowest crimes happened on December and February
